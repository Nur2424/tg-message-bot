## Load balanced personal dataset

```python
import pandas as pd

PATH = "../data/final/to_label_personal_balanced_1000.csv"
df = pd.read_csv(PATH)

if "label" not in df.columns:
    df["label"] = ""

df.head(3) 
```

## Interactive labeling loop

```python
label_map = {
    "+": "+",   # important
    "-": "-",   # skip
}

for idx in df.index:
    if isinstance(df.at[idx, "label"], str) and df.at[idx, "label"].strip() != "":
        continue

    print("\n" + "=" * 80)
    print(f"Index : {idx}")
    print(f"From  : {df.at[idx, 'from_name']}")
    print("-" * 80)
    print(df.at[idx, "text_clean"])
    print("-" * 80)
    print("Labels: [+] important | [-] skip ")

    key = input("Your choice (+ / -): ").strip()

    if key == "q":
        print("Stopping labeling.")
        break
    elif key == "s":
        continue
    elif key in label_map:
        df.at[idx, "label"] = label_map[key]
    else:
        print("Invalid input, skipping.") 
```

```python
df.to_csv("../data/final/labeled_1000_samples_manually.csv", index=False, encoding="utf-8")
print("Saved") 
```
